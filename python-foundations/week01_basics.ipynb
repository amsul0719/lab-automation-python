## 1) Lists, loops, and functions

```python
numbers = [10, 20, 30, 40, 50]

# Loop: print values
for n in numbers:
    print("Value:", n)

# Function: square a number
def sqare(n):
    return n*n

# Apply function to list (list comprehension)
squared_numbers = [sqare(n) for n in numbers]
squared_numbers
```

## 2) Plotting a simple transformation 

```python
import matplotlib.pyplot as plt

plt.plot(numbers, squared_numbers)
plt.title("Numbers vs Squared Numbers")
plt.xlabel("Number")
plt.ylabel("Square")
plt.show()
```

## Takeaway 

- Lists store ordered data that can be iterated over.
- Loops apply the same operation to each element.
- Functions encapsulate reusable logic.
- Simple plots help verify data transformations visually. 

## 3) Conditional logic (if/ elif/ else)

Conditional statements allow programs to make decisions based on values.

```python
x=15
if x>10: 
    print ("x is greater than 10")
else:
    print ("x is 10 or less")
```

```python
score = 72
if score >= 90:
    print("Excellent")
elif score >= 60:
    print("Pass")
else: 
    print("Fail")
```

### Classifying values using loops and conditions 

```python
numbers = [ 5, 12, 27, 60, 102]

for n in numbers: 
    if n < 10:
        print(n, "-> small")
    elif n <= 50:
        print (n, "-> medium")
    else: 
        print (n, "-> large")
```

### Takeaway 
- if/ elif/ else enables branching logic.
- Conditions can be combined with loops to classify data.
- This pattern is common in data filtering and QC steps.
